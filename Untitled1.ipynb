```python
import pandas as pd
import numpy as np
import seaborn as sns 
import warnings 
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Demand .csv")
```

```python
data.head()
```

```python
data.dtypes
```

```python
print(data.isnull().sum())
```

```python
def extract_cat_num(df):
    cat_col=[col for col in df.columns if df[col].dtype=='object']
    num_col=[col for col in df.columns if df[col].dtype!='object']
    return cat_col,num_col
```

```python
cat_col,num_col=extract_cat_num(data)
```

```python
for col in cat_col:
    print('{} has {} values '.format(col,data[col].unique()))
    print('\n')
```

```python
data=data.replace(to_replace = '- ', value = '0.0')
data=data.replace(to_replace = '-', value = '0.0')
```

```python
data=data.fillna(0)
```

```python
for i in range(len(cat_col)) :
    data[cat_col[i]] = data[cat_col[i]].astype(float)
```

```python
data.dtypes
```

```python
df=data
plt.rcParams["figure.figsize"] = (30,10)
# multiple line plots
plt.plot( 'Year', 'Total_capacity', data=df, marker='', markerfacecolor='blue', markersize=12, color='red', linewidth=4)
plt.plot( 'Year', 'Total_Used', data=df, marker='', color='black', linewidth=2,linestyle='dashed')



# show legend
plt.legend()

# show graph
plt.show()
```

```python
plt.plot( 'Year', 'Industrial_Use', data=df, marker='', markerfacecolor='blue', markersize=12, color='red', linewidth=4)
plt.plot( 'Year', 'Domestic_Use', data=df, marker='', color='black', linewidth=2,linestyle='dashed')



# show legend
plt.legend()

# show graph
plt.show()
```
