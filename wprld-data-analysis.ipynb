```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.rcParams["figure.figsize"] = (30,10)
```

```python
data=pd.read_csv("elec-fossil-nuclear-renewables.csv")
```

```python
data.head()
```

```python
data1=pd.DataFrame().assign(Ren=data['Sum of Electricity from renewables (TWh)'],Fosil=data['Sum of Electricity from fossil fuels (TWh)'],nuc=data['Sum of Electricity from nuclear (TWh)'])
```

```python
data_perc = data1.divide(data1.sum(axis=1), axis=0)
```

```python
data_perc
```

```python
# Make the plot
plt.stackplot(data.Year,  data_perc["Fosil"],  data_perc["nuc"],  data_perc["Ren"], labels=['Non_Renewable','Nuclaer','Green(Renewable)'])
plt.legend(loc='upper left')
plt.margins(0,0)
plt.title('Power Generation Over Years')
plt.show()
```

```python
df=data
plt.rcParams["figure.figsize"] = (30,10)
# multiple line plots
plt.plot( 'Year', 'Sum of Electricity from fossil fuels (TWh)', data=data, marker='', markerfacecolor='black', markersize=12, color='red', label="Oil")
plt.plot( 'Year', 'Sum of Electricity from renewables (TWh)', data=data, marker='', color='Green', linewidth=2,linestyle='dashed',label="Oil")
plt.plot( 'Year', 'Sum of Electricity from nuclear (TWh)', data=data, marker='', color='red', linewidth=2, linestyle='dashed', label="Oil")

# show legend
plt.legend()

# show graph
plt.show()
```
